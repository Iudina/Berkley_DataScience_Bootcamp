```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
data = pd.merge(ride_data, city_data, on="city", how="left")

# Display the data table for preview
data.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
grouped = data.groupby(['type', 'city'])

# Create DataFrame for the plot
plot_df = pd.DataFrame({"average fare": grouped["fare"].mean(), "driver count": grouped["driver_count"].mean(),
                    "total rides": grouped["fare"].count()})


plot_df.head()
```

```python
# Build the scatter plots for each city types
plt.scatter(plot_df.loc["Rural","total rides"], plot_df.loc["Rural", "average fare"], marker="o", 
            facecolors="gold", edgecolors="black", s= plot_df.loc["Rural", "driver count"] *10, alpha = 0.8, 
            label = "Rural")

plt.scatter(plot_df.loc["Urban","total rides"], plot_df.loc["Urban", "average fare"], marker="o", 
            facecolors="Salmon", edgecolors="black", s= plot_df.loc["Urban", "driver count"] *10, alpha = 0.8, 
            label = "Urban", )

plt.scatter(plot_df.loc["Suburban","total rides"], plot_df.loc["Suburban", "average fare"], marker="o", 
            facecolors="skyblue", edgecolors="black", s= plot_df.loc["Suburban", "driver count"] *10, alpha = 0.8, 
            label = "Subrban")

# Incorporate the other graph properties
plt.grid()
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total number of Rides (per city)")
plt.ylabel("Average Fare ($)")

# Create a legend
plt.legend(loc=1, markerscale = 0.5, title = "City Types", )

# Incorporate a text label regarding circle size
plt.text(42, 35, "Note:\nsize correlates with driver count per city")
plt.show()

# Save Figure
```

## Total Fares by City Type

```python
# Calculate Type Percents
grouped = data.groupby(['type'])
fares = grouped['fare'].sum()
percent_of_fares = [round(x/sum(fares)*100, 1) for x in fares]

# Build Pie Chart
slice_names = ["Rural","Suburban", "Urban"]
slice_colors = ["gold", "skyblue", "salmon"]
explode_index = (0, 0, 0.1)
plt.pie(percent_of_fares, explode=explode_index, labels=slice_names, colors=slice_colors, 
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Fares by City Type")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
rides = grouped['ride_id'].count()
percent_of_rides = [round(x/sum(rides)*100, 1) for x in rides]

# Build Pie Chart
slice_names = ["Rural","Suburban", "Urban"]
slice_colors = ["gold", "skyblue", "salmon"]
explode_index = (0, 0, 0.1)
plt.pie(percent_of_rides, explode=explode_index, labels=slice_names, colors=slice_colors, 
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("% of Total Rides by City Type")
# Save Figure
plt.show()
```
